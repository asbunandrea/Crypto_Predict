```python
# Import libraries and dependencies
import pandas as pd
import requests
import json
import datetime
```

```python
# Declare `url` variables
alternative_me = "https://api.alternative.me/fng/?limit=720"
```

```python
# Execute `GET` request with url
response_data = requests.get(alternative_me)
```

```python
 # Print `response_data variable`
print(response_data)
```

```python
# Store response using `content` attribute
response_content = response_data.json()
```

```python
print(response_content)
```

```python
# Convert to DataFrame
fng_df = pd.DataFrame(response_content["data"])

# Drop "time_until_update" column
fng_df = fng_df.drop(columns=["time_until_update"])
```

```python
# Preview
fng_df
```

```python
# Reorder columns
fng_df = fng_df[["timestamp","value","value_classification"]]
# Rename columns
fng_df = fng_df.rename(columns={"timestamp":"date"})
fng_df
```

```python
# Convert timestamp to date
fng_df.date = pd.to_datetime(fng_df.date, unit='s')
fng_df
```

```python
# convert to csv
fng_df.to_csv(r"fng.csv", index = False)
```
