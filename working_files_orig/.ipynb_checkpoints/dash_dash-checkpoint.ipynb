## Dash Data Preparation


```python
import numpy as np
import pandas as pd
import hvplot.pandas
from datetime import datetime, date
from functools import reduce
```

```python
# Load the fear and greed sentiment data for Bitcoin (only fear and greed index)
fng_df = pd.read_csv('fng.csv')
fng_df = fng_df.rename(columns={'date':'Date', 'value_classification':'fng_class'})
fng_df = fng_df.drop(columns=['value'])
mapping = {'Extreme Fear':'-1.0', 'Fear':'-0.5', 'Neutral':'0.0', 
        'Greed':'0.5', 'Extreme Greed':'1.0'}
fng_df['fng_class'] = fng_df['fng_class'].map(mapping)
fng_df['fng_class'] = pd.to_numeric(fng_df['fng_class'])
fng_df['fng_class'] = fng_df['fng_class'].astype(float)
fng_df.set_index(['Date'], inplace=True)
fng_df.head()
```

```python
# Load the vader sentiment data for Dash
dash_vader_df = pd.read_csv('dash_vader.csv', infer_datetime_format=True, parse_dates=True)
dash_vader_df["Publish Date"] = [date[0:10] for date in dash_vader_df["Publish Date"]]
dash_vader_df = dash_vader_df.drop(columns=["Post ID", "Title", "Url", "Author", "Score", "Total No. of Comments", "Permalink",
        "title_pos", "title_neu", "title_neg", "title_sent"], axis=1)
dash_vader_df.rename(columns = {'Publish Date':'Date', 'title_compound': 'vader_compound'}, inplace = True)
dash_vader_df['vader_compound'] = dash_vader_df['vader_compound'].astype(float).round(decimals=2)
dash_vader_df.set_index(['Date'], inplace=True)
dash_vader_df.head()
```

```python
# Load the historical closing prices for Dash
dash_close_df = pd.read_csv('dashprices.csv')
dash_close_df = dash_close_df.sort_index()
dash_close_df = dash_close_df.rename(columns={'date':'Date', 'close':'close_price'})
dash_close_df = dash_close_df.drop(["symbol", "volume"], axis=1)
dash_close_df.set_index(['Date'], inplace=True)
dash_close_df.head()
```

```python
# Load the exponential weighted moving average data for Dash
dash_ewm_df = pd.read_csv('dash_ema.csv')
dash_ewm_df = dash_ewm_df.sort_index()
dash_ewm_df = dash_ewm_df.rename(columns={'date':'Date', 'crossover_signal':'ewm_cross_signal'})
dash_ewm_df = dash_ewm_df.drop(["symbol", "close", "volume", "daily_return", "fast_close", "slow_close", "crossover_long", "crossover_short"], axis=1)
dash_ewm_df['ewm_cross_signal'] = dash_ewm_df['ewm_cross_signal'].astype(float)
dash_ewm_df.set_index(['Date'], inplace=True)
dash_ewm_df.head()
```

```python
# Load the money flow index for Dash
dash_mfi_df = pd.read_csv('mfi_dash.csv')
dash_mfi_df = dash_mfi_df.sort_index()
dash_mfi_df = dash_mfi_df.rename(columns={'date':'Date'})
dash_mfi_df = dash_mfi_df.drop(["symbol", "close", "volume", "Buy", "Sell"], axis=1)
dash_mfi_df['MFI'] = dash_mfi_df['MFI'].round(decimals=2)
dash_mfi_df.set_index(['Date'], inplace=True)
dash_mfi_df.head()
```

```python
# Load the money flow index for Dash
dash_google_df = pd.read_csv('dash_google.csv')
dash_google_df = dash_google_df.sort_index()
dash_google_df = dash_google_df.rename(columns={'Google Search Count':'google_search'})
dash_google_df['google_search'] = dash_google_df['google_search'].astype(float)
dash_google_df.set_index(['Date'], inplace=True)
dash_google_df.head()
```

```python
# MERGE the closing price, ewm, fng, mfi, and Google DataFrames into a single DataFrame
dash_dfs = [dash_close_df, dash_ewm_df, fng_df, dash_mfi_df, dash_google_df]
dash_data_df = reduce(lambda left, right: pd.merge(left, right, on='Date'), dash_dfs)
dash_df = dash_data_df.merge(dash_vader_df, on = 'Date', how = 'left')
#dash_df.replace(np.nan, 0)
dash_df['vader_compound'] = dash_df['vader_compound'].fillna(0)
dash_df.head()
```

```python
# MERGE the closing price, ewm, fng, mfi, and Google DataFrames into a single DataFrame
dash_dfs = [dash_close_df, dash_ewm_df, fng_df, dash_mfi_df, dash_google_df]
dash_data_df = reduce(lambda left, right: pd.merge(left, right, on='Date'), dash_dfs)
dash_df = dash_data_df.merge(dash_vader_df, on = 'Date', how = 'left')
#dash_df.replace(np.nan, 0)
dash_df['vader_compound'] = dash_df['vader_compound'].fillna(0)
dash_df.drop_duplicates()
dash_df.head()
```

```python
# Load the gas price data for Dash
dash_gas_df = pd.read_csv('dash_gas_price.csv')
dash_gas_df = dash_gas_df.sort_index()
dash_gas_df["f0_"] = [date[0:10] for date in dash_gas_df["f0_"]]
dash_gas_df = dash_gas_df.rename(columns={'f0_':'Date'})
dash_gas_df["std_gas"] = (dash_gas_df['f1_'] - dash_gas_df['f1_'].min())/(dash_gas_df['f1_'].max() - dash_gas_df['f1_'].min())
dash_gas_df.set_index(['Date'], inplace=True)
dash_gas_df = dash_gas_df.drop(columns=['f1_'], axis=1)
dash_gas_df.head()
```

```python
# Load the mining difficulty data for Dash
dash_diff_df = pd.read_csv('dash_difficulty.csv')
dash_diff_df = dash_diff_df.sort_index()
dash_diff_df["Date"] = [date[0:10] for date in dash_diff_df["Date"]]
dash_diff_df = dash_diff_df.rename(columns={'Date':'Date'})
dash_diff_df = dash_diff_df.sort_index()
dash_diff_df['Difficulty'] = dash_diff_df['Difficulty'].astype(float)
dash_diff_df["std_diff"] = (dash_diff_df['Difficulty'] - dash_diff_df['Difficulty'].min())/(dash_diff_df['Difficulty'].max() - dash_diff_df['Difficulty'].min())
dash_diff_df.set_index(['Date'], inplace=True)
dash_diff_df = dash_diff_df.drop(columns=['Difficulty'], axis=1)
dash_diff_df.head()
```

```python
# Load the actual Dash value (transfers) data for Dash (volume of transactions by day)
dash_value_df = pd.read_csv('actual_dash_value.csv') #, index_col="date", infer_datetime_format=True, parse_dates=True)
dash_value_df = dash_value_df.sort_index()
dash_value_df["f0_"] = [date[0:10] for date in dash_value_df["f0_"]]
dash_value_df = dash_value_df.rename(columns={'f0_':'Date'})
dash_value_df['f1_'] = dash_value_df['f1_'].astype(float)
dash_value_df["std_value"] = (dash_value_df['f1_'] - dash_value_df['f1_'].min())/(dash_value_df['f1_'].max() - dash_value_df['f1_'].min())
dash_value_df.set_index(['Date'], inplace=True)
dash_value_df = dash_value_df.drop(columns=['f1_'], axis=1)
dash_value_df.head()
```

```python
# MERGE the mining data into a single DataFrame
dfs = [dash_diff_df, dash_gas_df, dash_value_df]
dash_mining_df = reduce(lambda left, right: pd.merge(left, right, on='Date'), dfs)
dash_string_time = "2019-02-08"
dash_datetime = datetime.strptime(dash_string_time, "%Y-%m-%d")
dash_mining_df.rename(columns = {'f0_':'Date', 'std_diff':'mining_difficulty', 'std_gas': 'gas_price', 'std_value':'total_transfers' }, inplace = True)
dash_mining_df.head()
```

```python
# MERGE dash_mining_df with dash_df
dash_master_df = dash_df.merge(dash_mining_df, on = 'Date', how = 'left')
dash_master_df.head()
```

## LSTM Recurrent Neural Network for Historical and Predicted Closing Prices

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from numpy.random import seed
from tensorflow import random
from sklearn.preprocessing import MinMaxScaler
```

```python
# Set the random seed for reproducibility
seed(1)
random.set_seed(2)
```

```python
# Create function for features (X) and targeet(y)
def window_data(dash_combined, window, feature_col_number, target_col_number):
    X = []
    y = []
    for i in range(len(dash_combined) - window - 1):
        features = dash_combined.iloc[i:(i + window), feature_col_number]
        target = dash_combined.iloc[(i + window), target_col_number]
        X.append(features)
        y.append(target)
    return np.array(X), np.array(y).reshape(-1, 1)
```

```python
# Predict Closing Prices 
window_size = 10

# Column 0 is `Close Price`
feature_column = 0
target_column = 0
X, y = window_data(dash_combined, window_size, feature_column, target_column)
```

```python
# Use 70% of the data for training and the remainder for testing
split = int(0.7 * len(X))
X_train = X[: split]
X_test = X[split:]
y_train = y[: split]
y_test_scaler = y[split:]
```

```python
# Use the MinMaxScaler to scale data between 0 and 1.
scaler = MinMaxScaler()
scaler.fit(X)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
scaler.fit(y)
y_train = scaler.transform(y_train)
y_test_scaler = scaler.transform(y_test_scaler)
```

```python
# Reshape the features for the model
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

---

### Build and Train the Model

```python
# Define the LSTM RNN model.
model = Sequential()

# Initial model setup
number_units = 30
dropout_fraction = 0.2

# Layer 1
model.add(LSTM(
    units=number_units,
    return_sequences=True,             #carries the values forward
    input_shape=(X_train.shape[1], 1))
    )
model.add(Dropout(dropout_fraction))
# Layer 2
model.add(LSTM(units=number_units, return_sequences=True))
model.add(Dropout(dropout_fraction))
# Layer 3
model.add(LSTM(units=number_units))
model.add(Dropout(dropout_fraction))
# Output layer
model.add(Dense(1))
```

```python
# Compile the model
model.compile(optimizer="adam", loss="mean_squared_error")
```

```python
# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(X_train, y_train, epochs=50, shuffle=False, batch_size=100, verbose=0)
```
